```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

oo = pd.read_csv('olympics.csv', skiprows=4)
oo.head()
```

```python
mw = oo[(oo.Edition == 2008) & (oo.NOC == 'CHN')]
mw.Gender.value_counts().plot(kind='bar', color='blue', figsize=(5,5));
```

```python
sns.countplot(data=mw, x='Gender', palette='coolwarm');
```

```python
sns.countplot(data=mw, x='Medal', hue='Gender');
```

```python
sns.countplot(data=mw, x='Medal', hue='Gender', palette='coolwarm', order=['Gold', 'Silver', 'Bronze']);
```
